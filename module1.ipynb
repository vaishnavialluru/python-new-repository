```python
#strong number
number=int(input("enter a number: "))
temp=n
sum=0
while(number !=0):
  r=number%10
  f=1
  while(r !=1):
    f=f*r
    r=r-1
  sum=sum+f
  number=number//10
if(sum==temp):
  print("it is a strong number")
else:
  print("it is not a strong number")
```

```python
# palindrome
a=int(input("enter a number: "))
temp=a
reverse=0
while(a!=0):
  r=a%10
  reverse=reverse*10+r
  a=a//10
print(reverse)
if(reverse==temp):
  print("it is a palindrome")
else:
  print("it is not a palindome")
```
